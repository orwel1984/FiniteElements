## Lagrange Polynomials

Genral Formulae for N gird points, starting for index 0 to (n-1): 

$ L_j^{(n-1)}(x) = \prod_{\substack{i=0 \\ i \ne j}}^{n-1} \frac{x - x_i}{x_j - x_i} $

Degree n=1
N=2 grid points 

$
L_0^{(1)}(x) = \prod_{\substack{i=0 \\ i \ne 0}}^{1} \frac{x - x_i}{x_0 - x_i} = \frac{x - x_1}{x_0 - x_1}
$

$
L_1^{(1)}(x) = \prod_{\substack{i=0 \\ i \ne 1}}^{1} \frac{x - x_i}{x_1 - x_i} = \frac{x - x_0}{x_1 - x_0}
$


Degree n=2
N=3 grid points 

$
L_0^{(2)}(x) = \prod_{\substack{i=0 \\ i \ne 0}}^{2} \frac{x - x_i}{x_0 - x_i} 
= \frac{x - x_1}{x_0 - x_1} \cdot \frac{x - x_2}{x_0 - x_2}
$

$
L_1^{(2)}(x) = \prod_{\substack{i=0 \\ i \ne 1}}^{2} \frac{x - x_i}{x_1 - x_i} 
= \frac{x - x_0}{x_1 - x_0} \cdot \frac{x - x_2}{x_1 - x_2}
$

$
L_2^{(2)}(x) = \prod_{\substack{i=0 \\ i \ne 2}}^{2} \frac{x - x_i}{x_2 - x_i} 
= \frac{x - x_0}{x_2 - x_0} \cdot \frac{x - x_1}{x_2 - x_1}
$

## Calculating Shape functions for Quadrilateral Elements

Reference square $[\xi_0, \xi_1] *[\eta_0, \eta_1] $= [0,1] x [0,1]:

    η
    ^
    |
     4----3
     |    |
     |    |
     1----2 
    --> ξ



General 1D Lagrange polynomials:

For ξ-direction $ [\xi_0, \xi_1] = [0, 1] $ :

$L_0^{(1)}(\xi) = \frac{\xi - \xi_1}{\xi_0 - \xi_1} = 1 - \xi$

$L_1^{(1)}(\xi) = \frac{\xi - \xi_0}{\xi_1 - \xi_0} = \xi$

Thus:
$
L_0(\xi) = 1 - \xi, \quad L_1(\xi) = \xi
$


For η-direction $ [\eta_0, \eta_1] = [0, 1] $:  

$ L_0^{(1)}(\eta) = \frac{\eta - \eta_1}{\eta_0 - \eta_1} = 1 - \eta $

$ L_1^{(1)}(\eta) = \frac{\eta - \eta_0}{\eta_1 - \eta_0} = \eta $

Thus:
$
L_0(\eta) = 1 - \eta, \quad L_1(\eta) = \eta
$

| Shape function | $L_i^{(1)}(\xi)$ | $L_j^{(1)}(\eta)$ | Formula |
|----------------|------------|------------|---------|
| $\phi_1(\xi,\eta) = L_0(\xi).L_0(\eta)$ | $1-\xi$ | $1-\eta$ | $(1-\xi)(1-\eta)$ |
| $\phi_2(\xi,\eta) = L_1(\xi).L_0(\eta)$ | $\xi$    | $1-\eta$ | $\xi(1-\eta)$ |
| $\phi_3(\xi,\eta) = L_1(\xi).L_1(\eta)$ | $\xi$    | $\eta$   | $\xi \eta$ |
| $\phi_4(\xi,\eta) = L_0(\xi).L_1(\eta)$ | $1-\xi$ | $\eta$   | $(1-\xi)\eta$ |

Note: Each shape function value above is some product of both  $\xi, \eta$

| Shape function | $\frac{\partial \phi_i}{\partial \xi}$ | $\frac{\partial \phi_i}{\partial \eta}$ |
|----------------|---------------------------------------|---------------------------------------|
| $\nabla\phi_1$ | $-(1-\eta)$ | $-(1-\xi)$ |
| $\nabla\phi_2$ | $1-\eta$    | $-\xi$      |
| $\nabla\phi_3$ | $\eta$      | $\xi$       |
| $\nabla\phi_4$ | $-\eta$     | $1-\xi$     |


Each element of the Jacobian Matrix is a dot product of two 4 by 1 vectors:  

$
\frac{\partial x}{\partial \xi}=
\begin{bmatrix}
x_1 & x_2 & x_3 & x_4
\end{bmatrix}
\begin{bmatrix}
\dfrac{\partial \phi_1}{\partial \xi} \\
\dfrac{\partial \phi_2}{\partial \xi} \\
\dfrac{\partial \phi_3}{\partial \xi} \\
\dfrac{\partial \phi_4}{\partial \xi}
\end{bmatrix}
$

$
\frac{\partial y}{\partial \eta}=
\begin{bmatrix}
y_1 & y_2 & y_3 & y_4
\end{bmatrix}
\begin{bmatrix}
\dfrac{\partial \phi_1}{\partial \eta} \\
\dfrac{\partial \phi_2}{\partial \eta} \\
\dfrac{\partial \phi_3}{\partial \eta} \\
\dfrac{\partial \phi_4}{\partial \eta}
\end{bmatrix}
$


## Shape Functions for 1D Case

Shape functions are usually defined for 1D by stitching together two Lagrange Polynomials over adjacent elements to create  a Hat Function .<br>
This is shown as $\phi_i(x)$ for element i defined over $ [x_i , x_{i+1}]$ defined below:

$
\phi_i(x) = \begin{cases}
   L_i^{(1)}(x) = \frac{x - x_{i-1}}{x_i - x_{i-1}} = \frac{x - x_{i-1}}{h}  &\text{if } x \in [x_{i-1}, x_i ] 
   \\ 
   L_i^{(1)}(x) = \frac{x - x_{i+1}}{x_i - x_{i+1}} = \frac{x - x_{i+1}}{-h}  &\text{if } x \in [x_i , x_{i+1}] 
   \\
   0 &\text{else }
\end{cases}
$

$
\phi_i(x) =
\begin{cases}
\dfrac{x - x_{i-1}}{h}, & x \in [x_{i-1}, x_i], \\[1ex]
\dfrac{x_{i+1} - x}{h}, & x \in [x_i, x_{i+1}], \\[1ex]
0, & \text{otherwise},
\end{cases}
$

$
\phi_i(x) = L_i^{(1)}(x) :=
\begin{cases}
\dfrac{x - x_{i-1}}{x_i - x_{i-1}}, & x \in [x_{i-1}, x_i], \\[1ex]
\dfrac{x_{i+1} - x}{x_{i+1} - x_i}, & x \in [x_i, x_{i+1}], \\[1ex]
0, & \text{otherwise}.
\end{cases}
$

where h is nodal distances over the mesh points.

## Important Matrix operations  in Python / Numpy

### Matrix Basics in Python

#### Tip 1. ALWAYS use np.array[ ] for Linear Algebra / Maths

```python
import numpy as np

# Example 1 : BAD
B = [[1, 2, 3],   # row 0
     [4, 5, 6]]   # row 1   
print(type(B))  # Output: <class 'list'>
print(B[0][1])  # Access element at row 0, column 1 --> Output: 2   

# Example 2 : GOOD
# Use numpy arrays, more efficient, designed for numerical computations
A = np.array([
    [1, 2, 3],   # row 0
    [4, 5, 6]    # row 1
])
print(type(A))      # Output: <class 'numpy.ndarray'>
print(A[0, 1])       # Access element at row 0, column
```

```python
# Example 3 : Matrix of shape (2,3)
A = np.array([
    [1, 2, 3],   # row 0
    [4, 5, 6]    # row 1
])

print(A.shape) # Output (rows, columns) = (2, 3) 
print(A)
```

```python
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

print(A + B)      # [5 7 9]
print(A * 2)      # [2 4 6], broadcasting

# term-wise multiplication, not matrix multiplication
print(A * B)      # [4 10 18]

# For matrix multiplication, use np.dot or @ operator
np.dot(A, B)      # matrix multiply

# More operations
I = np.eye(3)  # identity matrix 3x3

A  = np.random.rand(2, 3)  
B  = np.random.rand(3, 2)  
C  = A @ B  # matrix multiplication

print(C)

np.linalg.inv(I)  # inverse

np.sum(A, axis=0) # sum along columns
```

## Code - Assemble Stiffness Matrices (non-vectorized)

```python
from scipy.sparse import coo_matrix
from scipy.sparse import coo_matrix
import numpy as np

from scipy.sparse import coo_matrix
from scipy.sparse import coo_matrix
import numpy as np

def Assemble(S, F, M, nodes, E, N):
    nL = nodes.shape[1]  # Nodes per Element

    # Initialize sparse matrices using LIL format for efficient assembly
    Sg = lil_matrix((N, N))
    Mg = lil_matrix((N, N))
    Fg = np.zeros(N)
    
    for i in range(E):  # Loop over all elements E (Quads or Triangles)

        for idx_i, i_G in enumerate(nodes[i]):      # Loop over nodes of  Element i
            Fg[i_G] += F[idx_i, 0, i] 

            for idx_J, j_G in enumerate(nodes[i]):  # Loop over nodes of  Element i
                Sg[i_G, j_G] += S[idx_i, idx_J, i]
                Mg[i_G, j_G] += M[idx_i, idx_J, i]

            # Why Loop twice over element nodes?
            #  Because Each S, M are nL x nL matrices per element

    # Convert to CSR for fast arithmetic / solving
    return Sg.tocsr(), Fg, Mg.tocsr()

```
